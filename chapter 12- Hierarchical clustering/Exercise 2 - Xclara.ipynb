```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('xclara.csv', sep = ",")
data.shape
```

```python
data.head()
```

```python
data.info()
```

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,5))
plt.title("Customer dendograms")
dend = shc.dendrogram(shc.linkage(data, method='ward'))
```

```python
#chia thanh 3 nhom
from sklearn.cluster import AgglomerativeClustering
cluster = AgglomerativeClustering(n_clusters = 3, affinity='euclidean', linkage='ward')
cluster.fit(data)
```

```python
cluster.labels_
```

```python
plt.scatter(data.V1, data.V2, c= cluster.labels_, cmap='rainbow')
plt.xlabel("V1")
plt.ylabel("V2")
plt.show()
```

```python
data['Group'] = cluster.labels_
data.head()
```

```python
group0 = data[data['Group'] == 0]
group0.describe()
```

```python
print("** V1: ")
print("mean: ",np.mean(group0[['V1']]))
print("min: ",np.min(group0[['V1']]))
print("max: ", np.max(group0[['V1']]))
print("** Spending Score: ")
print("mean: ",np.mean(group0[['V2']]))
print("min: ",np.min(group0[['V2']]))
print("max: ", np.max(group0[['V2']]))
```
